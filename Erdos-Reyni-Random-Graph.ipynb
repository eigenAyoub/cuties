```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def generate_graph(N, d): 
    p = d / (N-1)
    nodes = list(range(N))

    G = nx.Graph()
    G.add_nodes_from(nodes)
    
    edges = np.random.uniform(size= (N,N))

    ind = np.where(edges < p)
    x , y = ind[0], ind[1]

    for i,j in zip(x, y):
        G.add_edge(i,j)
        
    V = G.number_of_nodes()
    E = G.number_of_edges()
    
    stats = {"E": E,
             "V":  V,
             "Probb":      p,
             "d":d
            }
    
    return G, stats


def plot(G,st):
    ops = {
        'node_color': 'black',
        'edge_color': 'indigo',
        'node_size': 12,
        'width': 1,
    }
    
    fig, ax = plt.subplots(figsize=(14,7))
    
    nx.draw(G, **ops)

    plt.text(0.5, 1, 
             r'$\mathrm{E}[deg(i)] = %(d)0.2f  \quad (p = %(p)0.4f)$'\
             %{'d': st['d'], 'p':st['d']/(st['V']-1)},
             fontsize=15)
    fig.set_facecolor('lightcyan')
    plt.plot()
    plt.savefig(str(int(st['d']*10000))+".png", facecolor='lightcyan')
```

```python
n_of_vertices = 200
for d in np.linspace(0.6,1.4, 20):
    G, st = generate_graph(n_of_vertices,d)
    plot(G, st)
```
